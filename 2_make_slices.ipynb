This notebook makes slices out of 3D scans

```python
import os
import matplotlib.pyplot as plt
import numpy as np
import nrrd

for j in range(1, 21):
    file_path = f'KiTS\K{j}\K{j}.nrrd'
    file_path_seg = f'KiTS\K{j}\K{j}.seg.nrrd'

    data, header = nrrd.read(file_path)
    data_seg, header_seg = nrrd.read(file_path_seg)
    # Ensure the 'slices' directory exists
    output_dir = f'sliced/scan{j}'
    os.makedirs(output_dir, exist_ok=True)

    # Loop through each slice and save the segmented image
    for slice_index in range(data_seg.shape[2]):
        plt.imsave(os.path.join(output_dir, f'slice_seg{slice_index}.png'), data_seg[:, :, slice_index], cmap='gray')
        plt.imsave(os.path.join(output_dir, f'slice_{slice_index}.png'), data[:, :, slice_index], cmap='gray')

    print("Segmented slice images saved successfully in the 'slices' folder.")
```
